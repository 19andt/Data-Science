```python
%pylab inline
%matplotlib inline

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

pd.set_option('display.width', 5000) 
pd.set_option('display.max_columns', 60)

plt.rcParams['figure.figsize'] = (15, 5)
plt.style.use='default'
```

```python
total_list=pd.read_csv('Input/total_list.csv')
```

```python
total_list.pivot_table(values='births', index='year', columns='sex', aggfunc=sum).plot(title='Number of births per year', kind='line')
```

```python
def get_top1000(group):
    return group.sort_values(by='births', ascending=False)[:1000]

top1000_names=total_list.groupby(['year', 'sex']).apply(get_top1000)
```

```python
get_first_letter=lambda x: x[0]
first_letters=total_list.name.map(get_first_letter)
first_letters.name='First Letter'
```

```python
table=total_list.pivot_table(index=first_letters, values='births', columns=['sex', 'year'], aggfunc=sum)
subtable=table.reindex(columns=[1900, 1950, 2000], level='year')
```
