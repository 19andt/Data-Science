```python
from sklearn.base import BaseEstimator, TransformerMixin
from sklearn.utils import check_random_state

def inplace_relu(X):
    X[X < 0] = 0
    return X


def get_activation(name):
    if name is None:
        return lambda x: x
    elif name == 'tanh':
        return np.tanh
    elif name == 'relu':
        return inplace_relu
    else:
        raise ValueError("unknown activation: " + name)


class RandomEmbedder(BaseEstimator, TransformerMixin):
    
    def __init__(self, n_components=100, random_state=None, activation="tanh"):
        self.n_components = n_components
        self.random_state = random_state
        self.activation = activation

    def fit(self, X, y=None):
        n_samples, n_features = X.shape
        rng = check_random_state(self.random_state)
        self.components_ = rng.normal(size=(self.n_components, n_features))
        return self

    def transform(self, X, y=None):
        X_projected = np.dot(X, self.components_.T)
        return get_activation(self.activation)(X_projected)
```

```python
embedder = RandomEmbedder(n_components=2, activation="relu", random_state=0)
X_1 = embedder.fit_transform(X)
plt.scatter(X_1[:, 0], X_1[:, 1], c=y);
```

```python
from sklearn.pipeline import Pipeline

activation = "relu"

p = Pipeline([
    ("rnd", RandomEmbedder(n_components=10, activation=activation)),
    ("rnd2", RandomEmbedder(n_components=10, activation=activation)),
    ("rnd3", RandomEmbedder(n_components=10, activation=activation)),
    ("out", RandomEmbedder(n_components=2, activation=None)),
])
X_2 = p.fit_transform(X)
plt.scatter(X_2[:, 0], X_2[:, 1], c=y);
```

```python
from sklearn.base import BaseEstimator, TransformerMixin
from sklearn.linear_model import RidgeCV
from sklearn.utils import check_random_state


class AEELMTransformer(BaseEstimator, TransformerMixin):
    
    def __init__(self, n_components=100, activation='tanh',
                 alphas=np.logspace(-3, 3, 7), random_state=None):
        self.n_components = n_components
        self.alphas = alphas
        self.random_state = random_state
        self.activation = activation

    def fit(self, X, y=None):
        n_samples, n_features = X.shape
        rng = check_random_state(self.random_state)
        rnd_proj = rng.normal(size=(n_features, self.n_components))
        X_proj = get_activation(self.activation)(np.dot(X, rnd_proj))
        
        self.ridge_ = ridge = RidgeCV(alphas=self.alphas)
        ridge.fit(X_proj, X)
        
        self.components_ = ridge.coef_.T
        return self
        
    def transform(self, X, y=None):
        projected = np.dot(X, self.components_.T)
        return get_activation(self.activation)(projected)
```

```python
from sklearn.pipeline import Pipeline

activation = 'relu'
p = Pipeline([
    ("aeelm", AEELMTransformer(n_components=10, activation=activation)),
    ("aeelm2", AEELMTransformer(n_components=10, activation=activation)),
    ("aeelm3", AEELMTransformer(n_components=10, activation=activation)),
    ("out", RandomEmbedder(n_components=2, activation=None)),
])
X_3 = p.fit_transform(X)
plt.scatter(X_3[:, 0], X_3[:, 1], c=y);
```

```python
def shuffle_columns(X, copy=True, seed=None):
    rng = np.random.RandomState(seed)
    if copy:
        X = X.copy()
    for i in range(X.shape[1]):
        rng.shuffle(X[:, i])
    return X


def make_membership_pbm(X, seed=None):
    rng = np.random.RandomState(seed)
    n_samples, n_features = X.shape
    shuffle_idx = rng.permutation(2 * n_samples)
    
    X = np.vstack([X, shuffle_columns(X, seed=seed)])
    y = np.concatenate([np.ones(n_samples), -np.ones(n_samples)])
    return X[shuffle_idx], y[shuffle_idx]
```
