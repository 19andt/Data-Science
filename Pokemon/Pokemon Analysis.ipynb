# Pokemon data analysis

## 1. Loading the modules and data

**Loading the modules**

```python
%matplotlib inline

import warnings
warnings.filterwarnings("ignore")

import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

**Loading the data**

```python
data=pd.read_csv('Input/Pokemon.csv')
data.shape
```

```python
print(data.isnull().sum())
```

```python
data.head(3)
```

```python
data.sort_values('Total', ascending=False).head(5)
```

The best pokemon with overall rating is *RayquazaMega Rayquaza*, *MewtwoMega Mewtwo Y* and *MewtwoMega Mewtwo X*.

```python
data.sort_values('Attack', ascending=False).head(5)
```

The best pokemon for attack is *MewtwoMega Mewtwo X*.

```python
data.sort_values('Defense', ascending=False).head(5)
```

The best pokemon for attack is *Shuckle*, *SteelixMega Steelix* and *AggronMega Aggron*.

```python
data[data['Legendary']==True].sort_values('Total', ascending=False).head(10)
```

So the legendary pokemon are extremely powerfull than non-legendary pokemon.

```python
data[data['Legendary']==False].sort_values('Total', ascending=False).head(10)
```

```python
data[data['Legendary']==False].sort_values('Total', ascending=False).drop_duplicates(subset=['Generation'], keep='first').sort_values('Generation')[['Name', 'Generation', 'Total']]
```

The above table shows the best non-legendary pokemon across various generations.

```python
data[data['Legendary']==True].sort_values('Total', ascending=False).drop_duplicates(subset=['Generation'], keep='first').sort_values('Generation')[['Name', 'Generation', 'Total']]
```

The above table shows the best legendary pokemon across various generations.

```python
data.sort_values('Total').drop_duplicates(subset=['Type 1'], keep='last')[['Name', 'Type 1', 'Total']]
```

The above table shows the best pokemon different primary skill.

```python
data.sort_values('Total').drop_duplicates(subset=['Type 2'], keep='last').dropna()[['Name', 'Type 2', 'Total']]
```

The above table shows the best pokemon different secondary skill.

```python
data.sort_values('Attack').drop_duplicates(subset=['Type 1'], keep='last')[['Name', 'Type 1', 'Attack']]
```

The above table shows the best attacking pokemon among different skill set.

```python
data.sort_values('Defense').drop_duplicates(subset=['Type 1'], keep='last')[['Name', 'Type 1', 'Defense']]
```

The above table shows the best defensive pokemon among different skill set.

```python
data.sort_values('HP').drop_duplicates(subset=['Type 1'], keep='last')[['Name', 'Type 1', 'HP']]
```

The above table shows the pokemons with best *HP* among different skills.

```python
data.sort_values('Speed', ascending=False)[['Name', 'Speed']].head()
```

So the pokemon with greatest speed in *DeoxysSpeed Forme*.

```python
data['A+D']=data['Attack']+data['Sp. Atk']+data['Defense']+data['Sp. Def']
data.sort_values('A+D', ascending=False)[['Name', 'A+D']].head()
```

So the pokemons with best attack and defense are *KyogrePrimal Kyogre* and *GroudonPrimal Groudon*.

```python
plt.hist(data['Attack'], bins=range(0, 200, 20), histtype='bar',)
plt.axvline(data['Attack'].mean(),linestyle='dashed',color='red')
```

The mean attack of the pokemon is around 80.

```python
plt.hist(data['Defense'], bins=range(0, 250, 20), histtype='bar',)
plt.axvline(data['Defense'].mean(),linestyle='dashed',color='red')
```

The mean defense of the pokemon is around 75.

```python
explode=[]
for i in range(len(data['Type 1'].unique())):
    explode.append(0.05)
data['Type 1'].value_counts().plot(kind='pie', explode=explode, autopct='%1.1f%%')
fig=plt.gcf()
fig.set_size_inches(15, 15)
plt.show()
```

```python
sns.jointplot(x="Defense", y="Attack", data=data)
```

```python
sns.jointplot(x="HP", y="Attack", data=data)
```

```python
top_types=data['Type 1'].value_counts().head(10).reset_index()
df=data[data['Type 1'].isin(top_types['index'])]
sns.swarmplot(x='Type 1', y='Total', data=df, hue='Legendary')
plt.axhline(df['Total'].mean(),color='red',linestyle='dashed')
plt.show()
```

```python
df=data.drop(['#', 'Total', 'A+D'], axis=1)

plt.figure(figsize=(10,6))
sns.heatmap(df.corr(),annot=True)
plt.show()
```
