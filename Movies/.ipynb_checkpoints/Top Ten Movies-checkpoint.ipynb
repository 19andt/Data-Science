```python
# The usual preamble
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt

# Make the graphs a bit prettier, and bigger
pd.set_option('display.mpl_style', 'default')

# This is necessary to show lots of columns in pandas 0.12. 
# Not necessary in pandas 0.13.
pd.set_option('display.width', 5000) 
pd.set_option('display.max_columns', 60)

plt.rcParams['figure.figsize'] = (15, 5)
```

```python
ratings=pd.read_csv('ratings.csv')
movies=pd.read_csv('movies.csv')
users=pd.read_csv('users.csv')
```

```python
combined_data=pd.merge(pd.merge(ratings, users), movies)
combined_data.head()
```

```python
required_data=combined_data[['userId', 'movieId', 'rating', 'age', 'sex', 'title', 'genres']]
required_data[:5]
```

```python
average_ratings=required_data.groupby('movieId')['rating'].agg(['count', 'mean'])
average_ratings.head()
```

```python
top_ten_movies=average_ratings[average_ratings['count'] >= 50].sort_values('mean', ascending=False)[:10]
top_ten_movies
```

```python
top_ten_movies=pd.merge(top_ten_movies.reset_index(), movies)
top_ten_movies[['title', 'count', 'mean']]
```

# Gender Split

### Male Movie Rating

```python
average_ratings=required_data[required_data['sex']=='M'].groupby('movieId')['rating'].agg(['count', 'mean'])
average_ratings.head()
```

```python
male_top_ten_movies=average_ratings[average_ratings['count'] >= 50].sort_values('mean', ascending=False)[:10]
male_top_ten_movies
```

```python
male_top_ten_movies=pd.merge(male_top_ten_movies.reset_index(), movies)
male_top_ten_movies[['title', 'count', 'mean']]
```

### Female Movie Rating

```python
average_ratings=required_data[required_data['sex']=='F'].groupby('movieId')['rating'].agg(['count', 'mean'])
average_ratings.head()
```

```python
female_top_ten_movies=average_ratings[average_ratings['count'] >= 50].sort_values('mean', ascending=False)[:10]
female_top_ten_movies
```

```python
female_top_ten_movies=pd.merge(female_top_ten_movies.reset_index(), movies)
female_top_ten_movies[['title', 'count', 'mean']]
```
